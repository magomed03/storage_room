##Задание 3


```python
fig05 = plt.figure()
n = int(input())
```

```python
arr_points = []
for i in range(n):
    arr_points.append(tuple(map(int, input().split())))
print(arr_points)
```

```python
s = 0
for i in range(n):

    if (i == (n-1)):
        point = 0
    else:
        point = i+1
    s += 1/2 * np.linalg.det(np.matrix([[arr_points[i][0],arr_points[i][1]],[arr_points[point][0],arr_points[point][1]]]))
print(f"Площадь многоугольника равна: {s:.2f}")
```

```python
ax05 = fig05.add_subplot(1,1,1)
polygon = mpatches.Polygon(arr_points, edgecolor='#f20d75', facecolor='#0df28a')
ax05.add_patch(polygon)

x_coords, y_coords = zip(*arr_points)
ax05.plot(x_coords, y_coords, 'o', markersize=3, markerfacecolor='red', markeredgecolor='black')

ax05.set_xlim(0, 5)
ax05.set_ylim(0, 5)
ax05.set_aspect('equal')
fig05
```

```python
transform_matrix = np.array([[1, 0], [0, -1]])
arr_points_transformed = arr_points @ transform_matrix.T
s = 0
for i in range(n):

    if (i == (n-1)):
        point = 0
    else:
        point = i+1
    s += 1/2 * np.linalg.det(np.matrix([[arr_points_transformed[i][0],arr_points_transformed[i][1]],[arr_points_transformed[point][0],arr_points_transformed[point][1]]]))
print(f"Площадь отраженного многоугольника равна: {s:.2f}")
```

```python
fig06 = plt.figure()
ax06 =fig06.add_subplot(1,1,1)
polygon = mpatches.Polygon(arr_points_transformed, edgecolor='#f20d75', facecolor='#0df28a')
ax06.add_patch(polygon)
ax06.set_xlim(0, 5)
ax06.set_ylim(-5, 0)
ax06.set_aspect('equal')

x_coords, y_coords = zip(*arr_points_transformed)
ax06.plot(x_coords, y_coords, 'o', markersize=3, markerfacecolor='red', markeredgecolor='black')
```

##Задание 5

```python
centre = tuple(map(float, input("Введите координаты центра окружностей (x y): ").split()))
n = int(input("Введите количество сторон многоугольника: "))
side_length = float(input("Введите длину стороны многоугольника: "))

radius = side_length / (2 * np.sin(np.pi / n))  # радиус описанной окружности
r = radius * np.cos(np.pi / n)  # радиус вписанной окружности

fig08 = plt.figure()
ax08 = fig08.add_subplot(1, 1, 1)

circle_outside = plt.Circle(centre, radius, color='#0d75f2', fill=False, linestyle='dashed', linewidth=0.5)
plt.gca().add_patch(circle_outside)

circle_inside = plt.Circle(centre, r, color='#f20d75', fill=False, linestyle='dashed', linewidth=0.5)
plt.gca().add_patch(circle_inside)

angles = np.linspace(0, 2 * np.pi, n, endpoint=False)
vertices = np.array([[centre[0] + radius * np.cos(angle),
                      centre[1] + radius * np.sin(angle)] for angle in angles])

ax08.plot(vertices[:, 0], vertices[:, 1], 'o', color='#0df28a', markersize=4, label='Вершины')
ax08.plot(*np.append(vertices, [vertices[0]], axis=0).T, color='black', linewidth=1, label='Многоугольник')

for i, (x, y) in enumerate(vertices):
    ax08.text(x, y, f'P{i+1}', fontsize=6, ha='right', va='bottom', color='black')

ax08.set_xlim(0, 10)
ax08.set_ylim(0, 10)
ax08.grid()
ax08.set_title(f'Правильный {n}-угольник с описанными и вписанными окружностями')
ax08.legend(loc='upper right')

plt.show()
```
